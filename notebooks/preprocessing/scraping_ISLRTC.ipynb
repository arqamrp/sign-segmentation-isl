```python
import os
import sys
from pytube import YouTube, Playlist # to connect to youtube
from tqdm.notebook import tqdm
```

```python
playlist_link = 'https://www.youtube.com/playlist?list=PLFjydPMg4Dapq9vcdmGyHs8uJhiqMgUrX'
py = Playlist(playlist_link)
```

```python
os.mkdir('/kaggle/working/ISLRTC')

for i, video_link in tqdm(enumerate(py)):
    
    try:
        yt = YouTube(video_link)
        video_list = yt.streams.filter(only_video=True, mime_type='video/mp4').order_by('resolution').desc()
        vid = video_list[0]
        vid.download(output_path='/kaggle/working/ISLRTC', filename=f"video{i}.mp4")
    
    except:
        print(f"{i}= failed")

print(len(os.listdir('/kaggle/working/ISLRTC')))
```
